### Visualization vs. Statistics
Nominal  Categorical data가 있다.  
Ordinal  순서가 있으면 Ordinal하다.  
Interval  zero has no fixed meaning  
Ratio 0값이 기준점이 되게되면 Ratio의 성질을 가질수 있다라고 구분을 해놓습니다.

### Bar Chart
categorical 하고 ordinal한 데이터가 animal 데이터입니다  
기본적으로 Barchart for 1 D data  
animal data가 카테고리컬하고 오디널 (dog,rabbit,cat순서로)  
zero의 fixed meaning이 없고 0~1   

### Pie Chart
기본적으로 Pie Chart for 1D data  
Animal data가 Categorical하고, not ordinal 

### 히스토그램
기본적으로 히스토그램 for 1D data  
숫자만 있고, name은 없다. 순서도 없다  
Zero를 기준으로 비율을 그렸기때문에 Ratio의 개념이 있어요  
interval이 있다는 것은 각각 한칸한칸의 의미가 있다라는 식입니다.  
Numeric interval 0 has a special meaning(mean)

### Scatter plot
Scatter plot for 2D data  
숫자만 있고 name은 없다  
zero를 기준으로 발생을 점으로  
Numeric

### Line Plot
기본적으로 Line plot for 2D data  
숫자만 있고 name은 없다  
Zero를 기준으로 발생을 선으로  

### Box and whiskers
기본적으로 Box plot for 2D data  
anamal data가 Categorical, ordinal  
이름있는 Category가 순서로 정렬  
Weight data는 numeric data 0 has a special meaning (seight >0)  
숫자만 있고 name은 없다  

### HeatMap(matrix)

기본적으로 HeatMap for 2D data  
X축, Y축, 이름있는 Category가 순서로 정렬  
Zero의 fixed meaning이 없고, 0~1마리 같은 range 표현 없음  
0마리가 special meaning으로 볼 수 없다.   

### HeatMap(density or 2D Histogram)
기본적으로 HeatMap for 2D data  
숫자만 있고 name은 없다  
Zero를 기준으로 발생을 점으로

#### Side by side Barplot

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20.0, 10.0]
N=5
menMeans=(20,35,30,35,27)
womenMeans=(25,32,34,20,27)

ind=np.arange(N)
width=0.35
p1=plt.bar(ind-width/2,menMeans,width,color='#d62728')
p2=plt.bar(ind+width/2,womenMeans,width)
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind,('G1','G2','G3','G4','G5')) #그룹1, 그룹2, 그룹3, 그룹4, 그룹5
plt.yticks(np.arange(0,81,10))
plt.legend((p1[0],p2[0]),('Men','Women'))#그림에 사용된 방식 표현
```

##### Stacked Bar Plot

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20.0, 10.0]
N=5
menMeans=(20,35,30,35,27)
womenMeans=(25,32,34,20,27)

ind=np.arange(N)
width=0.35
p1=plt.bar(ind,menMeans,width,color='#d62728')
p2=plt.bar(ind,womenMeans,width,bottom=menMeans)
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind,('G1','G2','G3','G4','G5')) #그룹1, 그룹2, 그룹3, 그룹4, 그룹5
plt.yticks(np.arange(0,81,10))
plt.legend((p1,p2),('Men','Women'))#그림에 사용된 방식 표현
```

## SubPlot 화면나누기
Multiple plots with subplot!

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
```

```python
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)
plt.figure(1)
plt.subplot(211) # fig 2x1 matrix format  First
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')#k는 검정색 black

plt.subplot(212)# fig 2x1 matrix format  2nd 
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
```

```python
plt.figure(1)
plt.subplot(221)
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')
plt.subplot(122)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
```

##### Adding Text

```python
np.random.seed(19680801)
mu,sigma= 100,15
x=mu+sigma * np.random.randn(10000)

n,bins,patches=plt.hist(x,50,normed=1,facecolor='g',alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60,.025,r'$\mu=100,\ \sigma=15$')
plt.axis([40,160,0,0.03])
plt.grid(True)
```

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='white',color_codes=True)
```

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
sns.jointplot(x="sepal_length",y="sepal_width",data=iris,size=5)
```

##### iris data를 species 3가지로 구분

```python
sns.FacetGrid(iris,hue="species",size=5)\
.map(plt.scatter,"sepal_length","sepal_width")\
.add_legend()
```

```python
sns.boxplot(x='species',y='petal_length',data=iris)
```
